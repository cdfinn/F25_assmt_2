### Resampling data

- format a column as a date time
- set the index of our dataframe to this date time column
- resample our dataframe 

```python
data_311.dtypes
```

```python
data_311.head()
```

```python
data_311.set_index("Created Date")
```

```python
indexed_311_data=data_311.set_index("Created Date")
```

```python
tally_311=indexed_311_data.resample("M")["Unique Key"].count()
tally_311.head(20)
```

```python
indexed_311_data["Agency"].value_counts
```

### Plot your data 

```python
plt.plot(tally_311)
```

### Modifying  chart

```python
fig, ax = plt.subplots()
ax.plot(tally_311, color='red')
ax.set_xlabel('Years', fontsize = 8)
ax.set_ylabel('311 Calls', fontsize = 12)
ax.set_title('311 Complaints Near Austell Shelter')
fig.set_size_inches(6,3)

plt.savefig("../Output/chart.png")
```

### Analysis by Agency

```python
agency_indexed = data_311.set_index("Agency")
```

```python
agency_indexed.groupby(["Agency"]).agg({"Unique Key":'count'}).reset_index()
```

### Analysis by Street

```python
street_counts = data_311.groupby(["Street Name"]).agg({"Unique Key":'count'}).reset_index()
street_counts
```

```python
street_counts.plot(x='Street Name', y='Unique Key', kind='bar')
plt.xlabel('Street')
plt.ylabel('311 complaints')
plt.title('311 Complaints by Street')
plt.savefig("../Output/street_chart.png")
```
